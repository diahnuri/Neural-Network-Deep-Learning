```python
import matplotlib.pylab as plt
import numpy as np
```

```python
x = np.arange(-8, 8, 0.1)
print(x)
f = 1 / (1 + np.exp(-x))
```

```python
plt.plot(x,f)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title("the sigmoid function")
plt.show()
```

```python
w1 = 0.5
w2 = 1.0
w3 = 2.0
l1 = 'w1 = 0.5'
l2 = 'w2 = 0.1'
l3 = 'w3 = 2.0'
for w, l in [(w1,l1), (w2,l2), (w3,l3)]:
    f = 1 / (1 + np.exp(-x*w))
    plt.plot(x, f, label=l)
plt.xlabel('x')
plt.ylabel('h_w(x)')
plt.title('The effect of adjusting weights')
plt.legend(loc=2)
plt.show()
```

```python
w = 5
b1 = -8.0
b2 = 0.0
b3 = 8.0
l1 = 'b1 = -8.0'
l2 = 'b2 = 0.0'
l3 = 'b3 = 8.0'
for b,l in [(b1,l1), (b2,l2), (b3,l3)]:
    f = 1 / (1 + np.exp(-x*w+b))
    plt.plot(x,f,label=l)
plt.xlabel('x')
plt.ylabel('h_wb(x)')
plt.legend(loc=2)
plt.title('effect of bias adjustments')
plt.show()
```
